# Bivariate Analysis

```python
A
```

```python
A=A.drop(labels=["Manufacturer","Model","Make"],axis=1)
```

```python
#Separating continuous and categorical columns from the dataset
for i in A.columns:
    if(A[i].dtypes == 'object'):
        sb.boxplot(A[i],A.Width)#Boxplot between categorical and continuous columns
        plt.xlabel(i)
        plt.ylabel('Width')
        plt.show()
    else:
        plt.scatter(A[i],A.Width)# Scatter plot between continuous and continuous columns
        plt.xlabel('Width')
        plt.ylabel(i)
        plt.show()
```

```python
for i in A.columns:
    if(A[i].dtypes=="object"):
        sb.countplot(A.Origin,hue=A[i])#Countplot with hue between categorical and categorical columns
        plt.ylabel(i)
        plt.show()
```
